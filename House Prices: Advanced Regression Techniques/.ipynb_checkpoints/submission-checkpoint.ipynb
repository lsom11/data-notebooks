## Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# import seaborn as sns

pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('data/train.csv')

print(dataset.shape)
```

```python
dataset.head()
```

Look for following in EDA:

1. Missing Values
2. Numerical Variables
3. Distribution of Numerical Variables
4. Categorical Variables
5. Cardinality of Categorical Variables
6. Outliers
7. Relationship between independent and dependent features (SalePrice)

# Map Out Missing Values

```python
features_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1]

for feature in features_with_na:
    percent_missing = np.round(dataset[feature].isnull().mean(), 4)
    print(f"{feature} has {percent_missing} % of values missing")
```

# Find Relationship between missing values and the Sales Price

```python
for feature in features_with_na:
    data = dataset.copy()
    
    ## turn rows into 1 or 0 based on if it is null
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    ## mean sale price where information is missing or not
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

We can see that when there are NaN values, the median sales price is higher (in the majority of cases). We therefore need to see (in the feature engineering section) how we can replace these values.

There is a clear relationship between the missing values and the dependent variable.

# Numerical Variables

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
print(f"There are {len(numerical_features)} numerical features in the dataset")

dataset[numerical_features].head()
```

## Temporal Variables

Looking at this dataset we can see that there are 4 Year based variables.

```python
year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
for feature in year_features:
    print(feature, dataset[feature].unique())
```

### Let's try to understand if there is a relationship between these variables and sale price

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median Sale Price')
plt.title('House Price vs YearSold')
```

The price is decreasing.. which is weird because as the years pass the price should increase.. let's look at the other variables to see if there's a connection there.

```python
for feature in year_features:
    if feature != 'YrSold':
        data = dataset.copy()
        # we get the difference between the variable and the year the house was sold
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

This indicates that the less time between the year sold and the building or remodeling the higher the house sold for.

## Discrete Variables

```python
## 25 as threshold, less than that we can consider a discrete variable
discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_features + ['Id']]
print(f"There are {len(discrete_features)} discrete features in the dataset")
```

```python
discrete_features
```

```python
dataset[discrete_features].head()
```

Let's look for the relationship between our discrete features and the Sale Price

```python
for feature in discrete_features:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

## Continuous Variables

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_features + year_features + ['Id']]
print(f"There are {len(continuous_features)} continuous features in the dataset")
```

```python
continuous_features
```

```python
for feature in continuous_features:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

Not all continuous variables have a Guassian distribution, which shows this data is skewed and therefore normalization needs to happen.

```python
## logarithmic transformation to standardize data
for feature in continuous_features:
    data = dataset.copy()
    if 0 in data[feature].unique() or feature == 'SalePrice':
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Outliers

```python
for feature in continuous_features:
    data = dataset.copy()
    ## log 0 is undefined so we need to treat this
    if 0 in data[feature].unique() or feature == 'SalePrice':
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
